```python

import pandas as pd
from eda_setup import EDA

# 샘플 데이터프레임
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 25, 30, 35],
    'C': ['x', 'y', 'x', 'y', 'z']
})

# EDA 인스턴스 생성
eda = EDA(data)

# 메서드 호출
print(eda.summary())
print(eda.missing_values())
print(eda.data_types())
# eda.correlation_matrix()  # 이제 에러 없이 작동
# eda.missing_values_visual()
eda.plot_histograms()
eda.boxplot('A')
eda.plot_categorical('C')
eda.plot_histograms(sample_size=1000)  # 1000개 샘플로 히스토그램 생성
```
